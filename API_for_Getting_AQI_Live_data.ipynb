```python
def get_PM25_raw_from_subindex(aqi):
    if aqi <= 50:
        return aqi * 30 / 50
    elif aqi <= 100:
        return 30 + (aqi - 50) * 30 / 50
    elif aqi <= 200:
        return 60 + (aqi - 100) * 30 / 100
    elif aqi <= 300:
        return 90 + (aqi - 200) * 30 / 100
    elif aqi <= 400:
        return 120 + (aqi - 300) * 130 / 100
    elif aqi <= 500:
        return 250 + (aqi - 400) * 130 / 100
    else:
        return None
```

```python
import requests
from datetime import datetime

token = "dbe780a5142d9dc12434bf28545cb26eb7c27159"
station_id = 13738  # Perungudi station
url = f"https://api.waqi.info/feed/@{station_id}/?token={token}"
resp = requests.get(url)
resp.raise_for_status()
waqi_data = resp.json()

if waqi_data.get("status") == "ok":
    aqi = waqi_data["data"].get("aqi")
    iaqi = waqi_data["data"].get("iaqi", {})
    timestamp_str = waqi_data["data"].get("time", {}).get("s") # Extract timestamp string
    pollutants = {
        "PM2.5": iaqi.get("pm25", {}).get("v"),
        "PM10": iaqi.get("pm10", {}).get("v"),
        "NH3": iaqi.get("nh3", {}).get("v"),
        "SO2": iaqi.get("so2", {}).get("v"),
        "RH": iaqi.get("h", {}).get("v"), # Humidity
        "AQI": aqi # Add AQI
    }
    print("Live Air Quality - Perungudi, Chennai")
    if timestamp_str:
        # Parse the timestamp string and format it to show only the date
        try:
            date_obj = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
            formatted_date = date_obj.strftime('%Y-%m-%d')
            print(f"Date: {formatted_date}")
        except ValueError:
            print(f"Date: {timestamp_str}") # Print original string if parsing fails
    for k, v in pollutants.items():
        print(f"{k}: {v}")
else:
    print("Error:", waqi_data)
```

```python
from datetime import datetime

# Create a pandas DataFrame from the processed live data
live_df = pd.DataFrame([pollutants])

# Set the formatted date as the index
if formatted_date:
    live_df.index = [formatted_date]
    live_df.index.name = 'Date'
else:
    print("Warning: Date not found in live data. Using default index.")
    live_df.index.name = 'Date'

# Ensure columns match the structure of the existing data DataFrame
expected_columns = ['PM2.5', 'PM10', 'NH3', 'SO2', 'RH', 'AQI'] # Updated expected columns
for col in expected_columns:
    if col not in live_df.columns:
        live_df[col] = None # Add missing columns

# Select and reorder columns
live_df = live_df[expected_columns]

display(live_df.head())
```

```python
if not live_df.index[0] in data.index:
    # Append live_df to data if the date does not exist
    data = pd.concat([data, live_df])
    print("Live data appended to the historical data.")
else:
    # Inform the user that the date already exists
    print(f"Date {live_df.index[0]} already exists in the historical data. Live data not appended.")

display(data.tail())
```

```python
# Save the updated data DataFrame to the CSV file
data.to_csv('/content/drive/MyDrive/Tamil Nadu Case study/Data Set/perungudi_aqi_data.csv', index=True, mode='w')

print("Updated data saved to perungudi_aqi_data.csv")
```
